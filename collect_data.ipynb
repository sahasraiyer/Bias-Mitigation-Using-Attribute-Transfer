```python
import mediacloud.api, json, datetime
mc = mediacloud.api.MediaCloud('3853c1d6d5caee4837437baeb3dece9ed0369bceeea42349e88efda137b12460')

fetch_size = 500
conservative_stories = []
#conservative_stories channels 24669,28136,1092,19334
media_ids = [24669,1092]
last_processed_stories_id = 0
while len(conservative_stories) < 100:
    for m_ids in media_ids:
        fetched_stories = mc.storyList('media_id:'+str(m_ids), 
                                       solr_filter=mc.dates_as_query_clause(datetime.date(2015,1,1), datetime.date(2015,8,1)),
                                       last_processed_stories_id=last_processed_stories_id, rows= fetch_size)
        conservative_stories.extend(fetched_stories)
        if len( fetched_stories) < fetch_size:
            break
        last_processed_stories_id = conservative_stories[-1]['processed_stories_id']
print(json.dumps(conservative_stories))
```

```python
import mediacloud.api, json, datetime
mc = mediacloud.api.MediaCloud('6c5d05cf785dd55cb37687e0b95b5e0c8ee4ba0c5002d1f3d8fdef0e22d84b55')

fetch_size = 500
stories = []
#Neutral channels ,1150,1752,1091, 1755,7
media_ids = [1150]
last_processed_stories_id = 0
while len(stories) < 100:
    for m_ids in media_ids:
        fetched_stories = mc.storyList('media_id:'+str(m_ids), 
                                       solr_filter=mc.dates_as_query_clause(datetime.date(2016,11,1), datetime.date(2017,1,1)),
                                       last_processed_stories_id=last_processed_stories_id, rows= fetch_size)
        stories.extend(fetched_stories)
        if len( fetched_stories) < fetch_size:
            break
        last_processed_stories_id = stories[-1]['processed_stories_id']
print(json.dumps(stories))
```
